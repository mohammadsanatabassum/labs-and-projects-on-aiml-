singular value decomposition


```python
from skimage import data
import matplotlib.pyplot as plt
image = data.astronaut()

# Print some properties of the image
print("Data type:", image.dtype)
print("Min pixel value:", image.min())
print("Max pixel value:", image.max())
print("Shape:", image.shape)

# Display the image using matplotlib
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
import numpy as np
img = image/155.0
img = np.transpose(img, axes=(2,0,1))
U,S,V = np.linalg.svd(img)
print(U.shape, S.shape, V.shape)
```

```python
Sigma = np.zeros((3,512,512))
for i in range(3):
  np.fill_diagonal(Sigma[i,:,:], S[i,:])
print(Sigma.shape)
```

```python
k = 50
reconst = U @ Sigma[:,:,:k] @ V[:,:k,:]
reconst = np.transpose(reconst, axes=(1,2,0))
plt.imshow(reconst)
plt.show()
```

```python
reconst = U @ Sigma @ V
reconst = np.transpose(reconst, axes=(1,2,0))
plt.imshow(reconst)
plt.show()
```
